```python
import pandas as pd
import matplotlib.pyplot as plt

relax_types = ['model', 'model_relax_x'] #, 'model_relax_y', 'model_relax_w']
data = '/0613_30min'
records = ['_obj', '_runtime', '_mingap', '_bestbd']
```

```python
for relax_type in relax_types:
    for record in records:
        globals()[relax_type + record + '_df'] = pd.read_csv(relax_type + data + record + '.csv')
```
